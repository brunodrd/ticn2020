```python
def convert_decimal(dms, position):
    """
    Convertit une chaine de caractères représentant une coordonnée en DMS en flottant
    """
    
    if position == 'S' or position == 'W':
        signe = -1
    else:
        signe = 1
    
    val = float(dms)
    d = int(val / 100)
    m = int((val / 100 -d) * 100)
    s = round((val - int(val)) * 60)
    return (d + m / 60 + s / 3600) * signe
```

```python
convert_decimal('2117.7633', 'S')
```
